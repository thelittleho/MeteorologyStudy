````python
import seaborn as sns

data = sns.load_dataset('tips')

data
````

## Distribution Plot


---



### **Hist Plot**
   
히스토그램을 표시한다.

- 보통 하나 또는 둘의 변수를 가지고 있을 때 나타내는 전형적인 시각화 도구이다.



````python
sns.histplot(x=data['total_bill'])
# sns.displot(x=data['total_bill'], kind='hist')
````

````python
sns.histplot(x=data['tip'], y=data['total_bill'])
# sns.displot(x=data['total_bill'], kind='hist')
````

### **KDE plot**

````python
sns.kdeplot(x=data['total_bill'])
# sns.displot(x=df['total_bill'], kind='kde')
````

````python
sns.kdeplot(x= data['tip'], y=data['total_bill'])
# = sns.displot(x=df['total_bill'], y=df['tip'], kind='kde')
````

### **ECDF plot**
- 누적 분포를 그래프화 해 보여준다
- 관측치의 비율을 보여준다

````python
sns.ecdfplot(x=data['tip'])
````

### **Rug Plot**
- 눈금을 그려 표시를 해 분표도를 표기한다.
- 데이터마다 눈금을 표기해 만든 그래프이므로 보조그래프로 활용이 가능하다.

````python
sns.kdeplot(x=data["total_bill"])
sns.rugplot(x=data["total_bill"])
````
