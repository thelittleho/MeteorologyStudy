# Pandas란?
pandas는 파이썬의 대표적인 데이터 분석 라이브러리로   
데이터 조작, 분석, 시각화에 사용되는 라이브러리라고 생각하시면 될 듯 합니다.   
특히 2차원 데이터 구조인 DataFrame과 1차원 데이터 구조인 Series를 제공하여 테이블 형태의 데이터를 쉽게 다룰 수 있습니다.



```python
# Pandas의 기본 선언 방법
# pandas라는 이름 대신 pd라는 약칭으로 부르겠다는 의미이다.
import pandas as pd
```

# DataFrame
DataFrame 객체는 행과 열로 이루어진 2차원 데이터를 다루기 위한 객체입니다.   
- 열은 각각의 변수를 나타내고,    
- 행은 각각의 관측치를 나타냅니다.

## 리스트(list)를 이용한 DataFrame 생성

```python
data  = [['A', 1], ['B', 2], ['C', 3]]\
# columns가 없기 때문에 이를 꼭 지정해주어야만 한다.
df = pd.DataFrame(data, columns=['col1', 'col2'])

df
```

## 딕셔너리를 이용한 DataFrame 생성

```python
# 딕셔너리를 이용하는 방법이 상대적으로 깔끔하다...
data = {'col1' : ['a', 'b', 'c'], 'col2' : [1, 2, 3]}
df = pd.DataFrame(data)

df
```

## 특정 행 또는 열 출력

```python
# df의 col 출력
df['col1']
```

```python
# df의 0열 출력
df.loc[[0]]
```

## 행 또는 열 추가하기

```python
# 행 추가하기
df['Food'] = ['라면', '마라탕', '신전떡볶이']
# 열 추가하기
df['col3'] = [4, 5, 6]

df
```

## 행 또는 열 삭제하기

```python
# 행 삭제하기
df.drop([1], inplace=True)
# 열 삭제하기
df.drop('col3', axis=1, inplace=True)

df
```

# Series 객체
Series 객체는 인덱스와 값으로 이루어진 1차원 데이터를 다루기 위한 객체입니다. Series 객체는 DataFrame 객체에서 열을 선택하여 추출할 수 있습니다.

```python
data = [1, 2, 3]
s = pd.Series(data, index=['a', 'b', 'c'])

s
```

```python
data = {'a': 1, 'b':2}
s = pd.Series(data)

s
```
